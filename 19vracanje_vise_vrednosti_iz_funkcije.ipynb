## Враћање више вредности из функције

У примерима које смо до сада користили, функција је враћала једну вредност. У Пајтону нисмо ограничени на враћање само једне вредности. У Пајтону је могуће из функције вратити више вредности. <br>
Да би функција вратила више вредности потребно је након наредбе **`return`** навести све изразе које желимо да функција врати одвојене зарезима један од другог.
Пример изгледа наредбе за враћање више израза као резултате рада функције:

```Python
return izraz1, izraz,...
```
На пример погледајмо дефиницију функције **`obim_povrsina`**. Функција прихвата вредност полупречника и рчуна обим и површину круга. Након израчунавања, вредности обима и површине се враћају на место позива функције. 

````python
from math import pi

def obim_povrsina(pp):
    obim = 2 * pp * pi
    pov = pp ** 2 * pi
    
    return obim, pov
````

Када се позива ова функција у наредби доделе морамо користити две променљиве са леве стране знака једнако као што је приказано у следећем примеру:

````python
obim, pov = obim_povrsina(r)
````

Вредности наведене у наредби **`return`** са редом додељују променљивима које су наведене у позиву функције са леве стране знака једнакости.<br>
Када се заврши функција вредност првог израза после **`return`** наредбе додељује се првој променљивој у позиву функције и исто тако вредност другог израза након **`return`** наредбе додељује се другој променљивој у позиву функције.<br>
**Број променљивих са леве стране знака = приликом позива функције морају одговарати броју израза након `return` наредбе на крају функције.**<br>
Ако је број различит долази до грешке у програму.<br>


Пример комплетног програма:


````python
def main():
    r = float(input('Unesi poluprečnik: '))
    #poziva se funkcija, a promenljive obim i pov
    # će prihvatiti vrednosti koje se vrate iz funkcije
    obim, pov = obim_povrsina(r)
    
    print(f'Obim kruga je {obim:.2f}, a površina {pov:.2f}')

def obim_povrsina(pp):
    # na osnovu poluiprečnika (pp) dobijenog iz main funkcije
    # računa se obim i površina    
    obim = 2 * pp * pi
    pov = pp ** 2 * pi

    # vraćaju se obe vrednosti
    return obim, pov

from math import pi
main()
````

## О цикличним структурама

Ми смо у досадашњим програмима користили две програмске структуре: **линијску** и **разгранату**.

Карактеристике линијске структуре су следеће:

- свака наредба или израз се извршава само једном
- свака наредба или израз се мора извршити
- све наредбе се извршавају једна за другом одозго на доле

Друга врста структуре која се често користи је такозвана разграната структура.

Карактеристике разгранате структуре су следеће:

- одређена наредба или изрази ће се извршити ако је одређени услов задовољен или ако одређени услов није задовољен што зависи да ли се налази у **`if`**, **`elif`** или **`else`** делу.
- поједине наредбе се могу извршити али и не морају
- ако се извршавају, наредбе се извршавају само једном

Често се у програмима јавља потреба да се одређени делови програма (појединачне наредбе или изрази) понављају неколико пута. Користећи досадашње структуре наредбе које се морају понављати неколико пута морали бисмо писати на неколико места.


```python
ime = input('Unesi ime: ')
print(ime,'vidiš da nije teško biti programer.')
ime = input('Unesi ime: ')
print(ime,'vidiš da nije teško biti programer.')
ime = input('Unesi ime: ')
print(ime,'vidiš da nije teško biti programer.')
```

Такав начин решавања проблема има своје недостатке:

- програми су доста дужи
- неколико пута пишемо исте наредбе
- већа је шанса да се погреши код писања наредби

Да би се избегло непотребно понављање користе се такозване <span style="color:blue; font-weight:700">цикличне структуре</span> тј. трећа врста структура које ћемо упознати. Цикличне структуре се називају и <span style="color:blue; font-weight:700">петље</span>. 

**Коришћење петље у програму нам омогућава:**

- да програми буду краћи
- да мање времена утрошимо на писање програма
- да лакше можемо правити измене у програмима

Пошто наредбе које се понављају пишемо само на једном месту, то чини програм краћим и самим тим мање времена требa утрошити на писање програма. Aко је негде у програму потребно направити измену не морамо мењати на свим местима већ само на једном месту.

Подсетимо се једног проблема који смо решавали користећи разгранату структуру:<br>
потребно је да унесемо три броја и испишемо збир само једноцифрених бројева. Једно могуће решење тог проблема изгледало би овако:

- унети први број
- проверити да ли је једноцифрен - да ли је већи од нула и мање од 10
- ако јесте сабрати га са досадашњим збиром
- унети други број
- проверити да ли је једноцифрен
- ако јесте сабрати га са досадашњим збиром
- унети трећи број
- проверити да ли је једноцифрен
- ако јесте сабрати га са досадашњим збиром
- исписати збир једноцифрених или поруку да нема једноцифрених бројева

```python
zbir = 0
b1 = int(input('Unesi broj: '))
if 0 < b1 < 10:   #могли смо писати и на следећи начин if b1 > 0 and b1 < 10
    zbir = zbir + b1

b2 = int(input('Unesi broj: '))
if 0 < b2 < 10:
    zbir = zbir + b2

b3 = int(input('Unesi broj: '))
if 0 < b3 < 10:
    zbir = zbir + b3

if zbir > 0:
    print('Zbir jednocifrenih brojeva je:',zbir)
else:
    print('Niste uneli nijedan jednocifren broj.')
```

Пошто уносимо само три броја проблем можемо решити и овим кратким програмом али како би изгледао програм ако треба да унесемо 10, 50 или 100 бројева? Тада већ наилазимо на проблем. <br>
У горњем програму можемо уочити неколико наредби које се понављају. То су следеће наредбе:

<img src="slike/01ponavljanje.png">

Можемо уочити да понављамо наредбе за уношење бројева и наредбе које проверавају да ли је број једноцифрен. Ако јесте сабирамо га са тренутним збиром.<br>
Као што је већ речено ако треба унети и проверити 10, 50, 100 или више бројева тада би програм био знатно дужи и на много места би писали наредбе које се понављају.<br>
Да би то избегли можемо користити **петљу**. У следећем примеру за решење истог задатка користимо **`for петљу`**.

```python
zbir = 0

for i in (1,2,3):
    broj = int(input('Unesi broj: '))
    if 0 < broj < 10:
        zbir = zbir + broj

if zbir > 0:
    print('Zbir jednocifrenih brojeva je:',zbir)
else:
    print('Niste uneli nijedan jednocifren broj.')
```

У горњем примеру можемо видети да смо наредбу за уношење броја (<span style="color:red; font-weight:700">линија 4</span>), проверу да ли је он једноцифрен (<span style="color:red; font-weight:700">линија 5</span>) и сабирање са досадашњим збиром (<span style="color:red; font-weight:700">линија 6</span>) писали само на једном месту. Користећи <span style="color:blue; font-weight:700">for</span> петљу те наредбе ће се извршити 3 пута.

Исти пример можемо урадити и са другом врстом петљи - <span style="color:blue; font-weight:700">while</span> петљом.

```python
zbir = 0
i = 1
while i <= 3:
    broj = int(input('Unesi broj: '))
    if 0 < broj < 10:
        zbir = zbir + broj
    i = i + 1

if zbir > 0:
    print('Zbir jednocifrenih brojeva je:',zbir)
else:
    print('Niste uneli nijedan jednocifren broj.')
```

Дакле можемо видети да се у Пајтону користе две цикличне структуре:

- бројачка циклична структура или <span style="color:blue; font-weight:700">for</span> петља
- условна циклична структура или <span style="color:blue; font-weight:700">while</span> петља

**For петља**, чије ћемо коришћење детаљније упознати нешто касније, користи се у случајевима када тачно знамо колико пута желимо да одређени блок наредби понављамо. Пошто пре коришћења петље морамо знати колико ће се пута она извршавати, она се назива *бројачка петља*.

**While петљу** користимо када не знамо у напред колико ће петља трајати већ извршавање израза у телу петље петље зависи од испуњености одређеног логичког услова.

While петља је свеобухватнија петља што значи да све што се може урадити са for петљом може се урадити и са while петљом али обрнуто не важи.

For петља је за нијансу једноставнија и због тога се препоручује да све што се може урадити са for петљом уради помоћу ње.

Да погледамо сада како би урадили следећи пример. <br>
Задатак гласи да уносимо бројеве по жељи и да на крају испишемо збир свих унесених бројева. Овде ћемо само приказати могући начин рада са обе петље, а детаљније ћемо се наравно упознати у следећим лекцијама.

Kораци које ћемо урадити да би решили овај проблем су:

- Поставити почетну вредност за промењиву збир на нулу
- одлучити колико ћемо бројева уносити - **`n`**
- унутар петље **`n`** пута радити следеће
   - унети број
   - тај број сабрати са досадашњим збиром и то ће постати нови збир
- исписати укупан збир свих бројева

```python
zbir = 0
n = int(input('Koliko brojeva sabiraš: '))
for i in range(1,n + 1):
    broj = int(input(f'Unesi {i}. broj: '))
    zbir = zbir + broj

print(f'Zbir brojeva iznosi: {zbir}')
```

На почетку програма декларисали смо почетну вредност за промењиву **`zbir`** (<span style="color:red; font-weight:700">линијa 1</span>) и то ће нам бити променљива у којој ће се акумулирати сви бројеви које будемо уносили, тј. збир ће се увећавати за сваки унесени број.

У <span style="color:red; font-weight:700">линији 2</span> наводимо наредбу помоћу које одлучујемо колико ћемо бројева уносити. Ту вредност ћемо унети са тастатуре и доделити променљивој **`n`**.

Сада када знамо тачно колико ће се пута петља извршавати, а извршаваће се онолико пута колико ћемо бројева уносити, пишемо заглавље **`for`** петље.

<img src="slike/02zaglavljeForPetlje.png">

Промењива <span style="color:red; font-weight:700">i</span> у заглављу петље назива се **циклусна променљива**. Она преузима редом вредности из секвенце која је одређена **`range`** функцијом. Онолико колико има корака у **`range`** функцији или секвенци толико пута ће се наредбе и изрази у телу петља извршавати. <br>
Ако одредимо да ћемо уносити 5 бројева петља ће се извршавати 5 пута - <span style="color:red; font-weight:700">i</span> променљива ће добијати вредности од 1 до 6 (**али не и 6**) што значи 1, 2, 3, 4 и 5. <br>
У телу петље имамо две наредбе:
- Првом (<span style="color:red; font-weight:700">линијa 4</span>) уносимо број који желимо,
- другом (<span style="color:red; font-weight:700">линијa 5</span>) унесени број сабирамо са вредношћу досадашњег збира и то постаје нови збир. 

Када се заврши петља имаћемо укупан збир свих бројева и његову вредност исписујемо на екрану (<span style="color:red; font-weight:700">линијa 7</span>).

Исти проблем решићемо и помоћу **`while`** петље али ћемо користити нешто другачији начин. Рекли смо да **`while`** петљу користимо када не знамо тачно колико ће она имати корака. <br>Због тога ћемо овај проблем решити тако да ћемо у сваком кораку проверавати да ли има још бројева за уношење. Када више не буде било бројева за уношење (притиском на тастер **`Enter`** без уношења броја), тада ће се петља завршити и исписаћемо резултат.

```python
zbir = 0
dalje = 'da'
while dalje == 'da':
    broj = int(input('Unesi broj: '))
    zbir = zbir + broj
    dalje = input('Unosis jos brojeva da/ne: ')

print('Zbir brojeva je',zbir)
```

У првом реду (<span style="color:red; font-weight:700">линијa 1</span>) као и претходном примеру декларисали смо почетну вредност за промењиву **`zbir`** а то је нула. У другом реду (<span style="color:red; font-weight:700">линијa 2</span>) декларисали смо променљиву **`dalje`** коју ћемо користити приликом проверавања да ли имамо још бројева за уношење. На почетку смо јој доделили вредност **`da`**. </p><br>
Сада пишемо **`while`** петљу која ће се извршавати све док вредност променљиве **`dalje`** буде **`да`**. У телу петље се извршавају следеће наредбе:
- уносимо број (<span style="color:red; font-weight:700">линијa 4</span>) 
- унесени број додајемо промењивој збир (<span style="color:red; font-weight:700">линијa 5</span>) 
- постављамо питање кориснику да ли уноси још бројева и одговор додељујемо променљивој **`dalje`** (<span style="color:red; font-weight:700">линијa 6</span>).

**`While`** петља функционише тако да када се заврше наредбе у телу петље поново проверавамо постављени услов. Ако је вредност променљиве **`dalje`** опет **`da`** други пут улазимо у тело петље и извршавамо наредбе.

То значи да када се заврше све наредбе у телу петље враћамо се на заглавље и проверавамо услов. **Петља траје све док је тај услов тачан**. Када корисник унесе нешто што није **`da`** услов неће више бити тачан тј. вредност променљиве **`dalje`** неће бити једнака **`da`** и петља се завршава. 

Након петље исписујемо збир унесених бројева. (<span style="color:red; font-weight:700">линијa 8</span>)

- када је за решење одређеног проблема потребно писати наредбе које треба да се извршавају неколико пута користимо такозвану **цикличну структуру** или **петљу**.
- У пајтону постоје две врсте петљи: **`for`** петља и **`while`** петља
- **`for петља`** се назива још и *бројачка петља* и њу користимо у ситуацијама када тачно знамо колико ће се пута тело петље извршавати
- **`while петља`** се назива још и *условна петља* и њу користимо када унапред не знамо тачан број корака петље тј. колико пута ће се наредбе у телу петље извршити. Да ли ће се петља извршавати зависи од тога **да ли је услов у заглављу петље испуњен (тачан)**.
- након извршавање свих наредби у телу **`while`** петље поново проверавамо испуњеност услова. Ако је тачан опет се извршавају све наредбе у телу петље и тако све док у једном тренутку услов постане нетачан.
- **`while`** петља је моћнија тј. сви проблеми који се могу решити са **`for`** петљом могу се решити и са **`while`** петљом али проблеме које можемо решити са **`while`** петљом није увек могуће решити са **`for`** петљом.

## Враћање резултата рада функције

Функција која враћа резултат је тип функције која након завршетка наредби у телу функције враћа (прослеђује) резултат израчунавања у део програма из којег је позвана.
У првом делу лекција о функцијама видели смо начин коришћења такозваних **`void`** функција тј. функција које само извршавају наредбе у телу функције. Када се функција заврши програм се наставља од места са кога је позвана функција.
Функције са враћањем резултата су други тип функција. Врло су сличне **`void`** функцијама и имају следеће заједничке особине:
-	представљају групу наредби које обављају одређене задатак
-	када желите да се изврше морате их позвати

Функције са враћањем резултата имају још једну додатну особину: Када се заврши све наредбе у телу функције, враћају резултат рада функције у део програма са кога је позвана функција. Вредност која се враћа из функције може бити коришћена као било која друга вредност:
-	може бити додељена променљивој,
-	може бити приказана на екрану или
-	може бити коришћења у математичким изразима (ако је вредност нумеричка) итд,

Ове функције на крају тела функције имају израз ретурн који враћа вредност рада функције у део програма из кога је позвана функција.

### Дефинисање функција са враћањем вредности
Функције са враћањем вредности пишу се на исти начин као и **`void`** функције са једном разликом: Функције са враћањем вредности морају имати наредба **`return`** на крају тела функције.


Вредност која се враћа из функције може се даље у програму користити као било која друга вредност:
-	може се користити у другим израчунавањима
-	може се исписивати у **`print`** функцији
-	може се послати као вредност другој функцији…

Општи облик функција са враћањем вредности:
```Python
def ime_funkcije():
    naredba
    naredba
    itd...
    return vrednost
```

Једна од наредби у функцији мора бити наредба **`return`** која има следећи облик:
```Python
return izraz
```

Вредност израза који се наводи након речи **`return`** шаље се назад у део програма на место из кога је позвана функција. То може бити одређена вредност, променљива или израз који као резултат има одређену вредност.

Следи пример функције са враћањем резултата:

````python
def zbir_brojeva(br1, br2):
    rezultat = br1 + br2
    return rezultat
````

Намена ове функције јесте да прихвати две бројчане вредности као аргументе и враћа њихов збир. 
Прва наредба у телу функције додељује вредност израза **`br1 + br2`** променљивој **`rezultat`**. 
Након тога извршава се наредба **`return`** којом се завршава функција и враћа вредност променљиве **`rezultat`** у део програма са кога је позвана функција.

Следећи програм приказују употребу функције са враћањем резултата:

````python
# program koristi vrednost dobijenu iz funkcije

def main():
   # unosi se iznos zarade učenika
   zarada = int(input("Unesi zaradu: "))

   # unosi se iznos džeparca
   dzeparac = int(input("Unesi iznos džeparca: "))

   #prihvatanje ukupne vrednosti oba iznosa
   zajedno = ukupno(zarada, dzeparac)

   # prikazivanje ukupne vrednosti
   print(f'Stanje štednje: {zajedno} dinara.')

# funkcija prihvata dve vrednosti i vraća
# rezultat zbira te dve vrednosti

def ukupno(iznos1, iznos2):
    zbir = iznos1 + iznos2
    return zbir

# pozivanje main funkcije
main()
````

У **`main`** функцији у програм се преко тастатуре уносе две вредности (два броја) и додељује променљивима **`zarada`** и **`dzeparac`**. Наредба у **линији 11** позива функцију **`ukupno`** и прослеђује јој вредности променљивих **`zarada`** и **`dzeparac`** као аргументе.

Функција **`укупно`** сабира вредности параметара функције **`iznos1`** и **`iznos2`** њихов збир додељује променљивој **`zbir`**. 

Вредност која се враћа из функције се додељује променљивој **`zajedno`** (променљива којој се додељује вредност рада функције).

На следећој слици приказано је како су аргументи прослеђују функцији и како се вредност враћа из функције



### Други начин за враћање резултата

Погледајмо још једном претходни пример
```Python
def ukupno(iznos1, iznos2):
    zbir = iznos1 + iznos2
    return zbir
```
Можемо приметити да смо навели две наредбе:
1.	Вредности израза износ један плус износ два смо доделили променљивој збир
2.	Из функције се враћа вредност променљиве збир

Иако ова функција ради оно за шта је намењена она се може мало поједноставити. Говорећи о функцијама које враћај резултат рекли смо да функција може вратити вредност променљиве, логичку уредност или резултат математичког израза.
Пошто се може вратити резултат математичког израза који је наведен после наредбе **`return`**, функцију смо могли написати и на следећи начин:
```Python
def ukupno(iznos1, iznos2):
    return iznos1 + iznos2
```
Ова верзија функције не смешта резултат израза **`iznos1 + iznos2`** у променљиву. Уместо тога користи се могућност да се из функције враћа резултат математичког израза.

Тако ће се у наведеном случају прво сабрати вредности променљивих **`iznos1`** и **`iznos2`** и та вредност вратити из функције.
Вредност која се враћа из функције можемо као резултат приказати и у **`print`** функцији. У овом случају унутар **`print`** функције у којој се исписује порука, позваћемо функцију и она, када се заврши, на то место са ког је позвана исписаће резултат тј. израз који је наведен иза **`return`** наредбе о функцији.

У нашем примеру то би се могло написатина следећи начин:

```Python
print(f'Stanje štednje: {ukupno(zarada, dzeparac)} dinara.')
```

Комплетан програм изгледа овако:

````python
# program koristi vrednost dobijenu iz funkcije

def main():
   # unosi se iznos zarade učenika
   zarada = int(input("Unesi zaradu: "))

   # unosi se iznos džeparca
   dzeparac = int(input("Unesi iznos dzeparca: "))

   # prikazivanje ukupne vrednosti
   print(f'Stanje štednje: {ukupno(zarada, dzeparac)} dinara.')

# funkcija prihvata dve vrednosti i vraća rezultat
# zbira te dve vrednosti
def ukupno(iznos1, iznos2):
    return iznos1 + iznos2

# pozivanje main funkcije
main()
````

У **линији 11** наведена је **`print`** функција у којој се исписује порука и позива функција **`ukupno(zarada, dzeparac)`**
Када се функција изврши, резултат рада функције ће се вратити на место са кога је позвана тј исписати у **`print`** функцији.

### Враћање више вредности

У примерима које смо до сада користили, функција је враћала једну вредност. У Пајтону нисмо ограничени на враћање само једне вредности. У Пајтону је могуће из функције вратити више вредности. 

Да би функција вратила више вредности потребно је након наредбе **`return`** навести све изразе које желимо да функција врати одвојене зарезима један од другог.

Пример изгледа наредбе за враћање више израза као резултате рада функције:
```Python
return izraz1, izraz2, itd
```
На пример погледајмо дефиницију функције **`unesi?ime`**. Функција тражи од корисника да се унесе име и презиме. Унесене вредности са смештају у две локалне променљиве **`ime`** и **`preyime`**. Наредба **`return`** враћа вредности обе променљиве:

````python
def unesi_ime():
   # unosi se ime i prezime učenika
    ime = input("Unesite ime ucenika: ")
    prezime = input("Unesite prezime ucenika: ")

    # vraćaju se obe vrednosti
    return ime, prezime
````

Када се позива ова функција у наредби доделе морамо користити две променљиве са леве стране знака једнако као што је приказано у следећем примеру:

```Python
ime_ucenika, prezime_ucenika = unesi_ime()
```

Вредности наведене у наредби **`return`** се редом додељују променљивима које су наведене у позиву функције са леве стране знака једнакости.
Када се заврши функција вредност првог израза после **`return`** наредбе додељује се првој променљивој у позиву функције и исто тако вредност другог израза након **`return`** наредбе додељује се другој променљивој у позиву функције.

Број променљивих са леве стране знака = приликом позива функције морају одговарати броју израза након **`return`** наредбе на крају функције. Ако је број различит долази до грешке у програму.

Пример комплетног програма:

````python
def main():
    
    imeUcenika, prezimeUcenika = unesi_ime()
    
    print('\nPodaci o uceniku')
    print(f'Ime ucenika: {imeUcenika}')
    print(f'Prezime ucenika: {prezimeUcenika}')

def unesi_ime():
   # unosi se ime i prezime učenika
    ime = input('Unesite ime ucenika: ')
    prezime = input('Unesite prezime ucenika: ')

    # vraćaju se obe vrednosti
    return ime, prezime

main()
````

## Локалне променљиве у функцији

<div class="alert alert-block alert-info">
<strong>Укратко</strong><br>
Локалне променљиве су променљиве које се креирају унутар функције и не могу бити коришћене изван функције. Оне служе само за коришћење у наредбама унутар функције.
Различите функције могу имати истоимене променљиве због тога што функција не препознаје променљиве у другим функцијама.
</div>    

Честo се јавља потреба да у функцији користимо одређене променљиве. Сваки пут када креирате променљиву унутар функције ви креирате **локалну променљиву**. Локална променљива припада функцији у којој је креирана и само наредбе унутар функције могу да користе ту променљиву. Ако изван функције покушате приступити променљивој која је декларисана унутар функције доћи ће до грешке у програму.

```python
def veci_broj():
    broj1 = int(input('Unesi prvi broj: '))
    broj2 = int(input('Unesi drugi broj: '))
    if broj1 > broj2:
        print('Veći broj je', broj1)
    elif broj2 > broj1:
        print('Veći broj je', broj2)
    else:
        prin('Brojevi su isti')

print('Program ispisuje veći od dva broja.')
veci_broj()
zbir = broj1 + broj2
print('Zbir brojeva je', broj1 + broj2)    
```

Приликом извршавања горњег програма јавља се грешка због тога што за програм променљива  **`broj1`** није позната.

<strong>ВАЖНО</strong><br>
Локалне променљиве су променљиве које се креирају унутар функције и не могу бити коришћене изван функције. Локалне поменљиве се креирају и бришу сваки пут када се функција извршава и трају док траје функција.
</div>   

Уколико је потребно да се променљива користи у неколико функција, она се може дефинисати као глобална. Проенљива ће имати глобалну важност само ако у одређеној функији не постоји функција са истим именом.

**Коришћење глобалних променљивих у програмима треба избегавати пошто отежавају читљивост програма.**

Уколико је потребно у функцији користити вредност променљиве из друге функције тада се вредност променљиве прослеђује као аргумент (учићемо у следећој лекцији).

Коришћење глобалних променљивих је приказан у следећем програму

```python
def veci_broj():
    # дефинисање променљивих као глобалних
    global broj1, broj2
    
    broj1 = int(input('Unesi prvi broj: '))
    broj2 = int(input('Unesi drugi broj: '))
    if broj1 > broj2:
        print('Veći broj je', broj1)
    elif broj2 > broj1:
        print('Veći broj je', broj2)
    else:
        prin('Brojevi su isti')

print('Program ispisuje veći od dva broja.')
veci_broj()
zbir = broj1 + broj2
print('Zbir brojeva je', broj1 + broj2)    
```

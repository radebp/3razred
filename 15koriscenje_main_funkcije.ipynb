## Коришћење `main` функције

Погледајмо један пример програма који користи функције:

```python
def krug():
    from math import pi
    r = float(input('Unesi poluprečnik: '))
    obim = 2 * r * pi
    pov = r ** 2 * pi
    print(f'Obim kruga je {obim:.2f}, a površina {pov:.2f}')

def pravougaonik():
    a = float(input('Unesi dužinu stranice a: '))
    b = float(input('Unesi dužinu stranice b: '))
    obim = 2 * (a + b)
    pov = a * b
    print(f'Obim pravougaonika je {obim:.2f}, a površina {pov:.2f}')

fig = input('Izaberi figuru (krug ili pravougaonik): ')
if fig == 'krug':
    krug()
elif fig == 'pravougaonik':
    pravougaonik()
else:
    print('Pogrešna figura')
```

Када у програму користимо функције програм не почиње да се извршава од прве наредбе у листингу програма као што је то до сада био случај. Када програм дође до дефиниције функције он је смешта у меморију али је не извршава. У овом случају програм ће почети да се извршава од **линије 15**. <br>
Када изаберемо жељену фигуру, ако је у питању круг позива се функција за рачунање обима и површине круга (**линија 17**). Ако изаберемо правоугаоник позива се функција за израчунавање и исписивање обима и површине правоугаоника (**линија 19**).

Претходни програм има две функције али је наравно могуће имати много функција у програму. 
>**Честа пракса приликом писања програма јесте да се све потребне операције смештају у функције.**

Један од честих начина писања програма са функцијама је коришћење **`main`** (главне) функције која се позива када програм почне са извршавањем.

**`Main`** функција тада позива остале функције у програму када је потребно њихово извршавање.

У следећем примеру програм је  написан користећи две функције: функције за исписивање поруке и `main` функције.


```python
#Program ima dve funkcije
# Prvo definišemo main funkciju
def main():
    print('Posle ove rečenice izvršava se druga funkcija.')
    poruka()
    print('Ponovo smo u main fukciji')
    
# sledi definisanje sledeće funkcije
def poruka():
    print('Pozvali ste me da se izvršim!')
    print('I ova naredba je u funkciji "poruka()".')
    
# ovde počinje izvršavanje programa
# poziva se main funkcija
main()
```

Дефиниција **`main`** функције почиње од **линије 3** до **линије 6**. Дефиниција функције **`poruka`** почиње од **линије 9** до **линије 11**. Наредба у **линији 15** је прва наредба која се извршава у програму и она служи да позове или покрене **`main`** функцију.


Прва наредба у **`main`** функцији позива **`print`** функцију да се изврши (**линија 4**). Она приказује стринг “Posle ove naredbe izvršava se druga funkcija”. Након тога наредба у **линији 5** позива функцију **`poruka`**. Тада програм прелази на извршавање функције **`poruka`** која исписује 2 поруке на екрану.


Када се заврши функција **`poruka`** програм се враћа у **`main`** функцију и наставља са првом следећом наредбом после позива функцији **`poruka`**, а то је исписивање стринга "Ponovo smo u main funkciji".
![main_i_poruka3.png]

Када се испише порука "Ponovo smo u main funkciji" завршава се **`main`** функција и програм се враћа на место са кога је позвана **`main`** функција.
Пошто нема више наредби после позива **`main`** функцији програм се завршава

Када програм позива функцију обично се каже да се контрола извршавања програма пребацује на функцију. То значи да функција преузима даље извршавање програма. Када се изврше све наредбе у телу функције, програм се “враћа” на место са кога је функција позвана и прелази на прву следећу наредбу после позива функције или се програм завршава.

Следи пример програма који рачуна и исписује површину и обим правоугаоника или круга у коме су све наредбе смештене у три функције: **`main`**, **`krug`** и **`pravougaonik`**

```python
def krug():
    from math import pi
    r = float(input('Unesi poluprečnik: '))
    obim = 2 * r * pi
    pov = r ** 2 * pi
    print(f'Obim kruga je {obim:.2f}, a površina {pov:.2f}')

def pravougaonik():
    a = float(input('Unesi dužinu stranice a: '))
    b = float(input('Unesi dužinu stranice b: '))
    obim = 2 * (a + b)
    pov = a * b
    print(f'Obim pravougaonika je {obim:.2f}, a površina {pov:.2f}')

def main():
    fig = input('Izaberi figuru (krug ili pravougaonik): ')
    if fig == 'krug':
        krug()
    elif fig == 'pravougaonik':
        pravougaonik()
    else:
        print('Pogrešna figura')

main()
```

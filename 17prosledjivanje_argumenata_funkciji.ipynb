## Прослеђивање аргумената функцији

<span style="color:red; font-weight:bold">Аргумент</span> - податак који се прослеђује функцији када се функција позива.

<span style="color:red; font-weight:bold">Параметар</span> наводи се у заглављу функције и предтсвља променљиву која прима вредност аргумента који се прослеђује функцији.

Често је корисно не само позвати функцију већ јој послати одређени податак. Податак који се шаље функцији назива се **аргумент**. 

Функција користи аргументе за разна израчунавања или друге операције унутар функције.
Ако желите да функцији пошаљете аргументе када је позивате, морате у дефиницији функције навести један или више параметара тј. променљивих које ће да приме вредности које буду послате функцији.

У следећем примеру дефинисана је функције која има 1 параметар. **Параметар** је променљива **`broj`** која ће да прими вредност која буде послата функцији приликом позива функције.

```Python
def prikazi_kub(broj):
    rezultat = broj ** 3
    print(f'Kub broja {broj} iznosi {rezultat}')
```

У горњем примеру дефинисали смо функцију под именом **`prikazi_kub`**. Њен задатак је да прими број као аргумент и прикаже вредност куба тог броја. <br>
Ако погледамо заглавље функције приметићемо реч **`broj`** која се налази унутар заграда. То је име променљиве која је параметар функције. Та променљива ће примити вредност аргумента који ће се проследити функцији када функција буде позвана.

Следећи програм приказује функцију и комплетан програм


````python
# program demonstrira prosleđivanje argumenata funkciji

def main():
    vrednost = 5
    prikazi_kub(vrednost)

# funkcija prikazi_kub prihvata argument i prikazuje njegov kub
def prikazi_kub(broj):
    rezultat = broj ** 3
    print(f'Kub broja iznosi {rezultat}')

# pozivanje main funkcije
main()
````

Када се програм покрене позива се **`main`** функција у **линији 13**. Унутар **`main`** функције у **линији 4** креирана је локална променљива под именом **`vrednost`** и додељена јој је вредност **`5`**. Након тога у **линији 5** позива се функција **`prikazi_kub`**.

```Python
prikazi_kub(vrednost)
```

Можемо видети да се променљива **`vrednost`** налази унутар заграда. То значи да ће се вредност проследити као аргумент функцији **`prikazi_vrednost`** као што је приказано на следећој слици.



Када се изврши та наредба (позив функцији **`prikazi_kub`**) почиње са извршавањем функција **`prikazi_kub`** и параметру **`broj`** се додељена вредност променљиве **`vrednost`**.
То је приказано на следећој слици



Да анализирамо функцију **`prikazi_kub`**

Као што видимо параметарској променљивој **`broj`** биће додељена вредност када се позове функција и тада јој се прослеђује вредност аргумента приликом позива функцији. У горњем програму то се налази у **`линији 5`**. У **`линији 9`** вредности израза **`broj ** 3`** додељује се променљивој **`rezultat`**. Пошто је вредност променљиве **`vrednost`** једнака **`5`** овај израз додељује број **`125`** променљивој **`rezultat`**.
У **`линији 10`** приказује се вредност променљиве **`rezultat`**.
У следећем изразу приказано је како се функција **`prikazi_kub`** може позвати са одређеном вредношћу као аргументом

```Python
prikazi_kub(8)
```
Горња наредба позива функцију **`prikazi_kub`** и прослеђује јој **`8`** као аргумент који ће бити додељен параметру **`broj`** у функцији **`prikazi_kub`**.


У претходном програму, функцији **`prikazi_kub`** се увек прослеђује вредност 5. У следећем примеру корисник уноси жељени број и тај број прослеђује функцији **`prikazi_kub`** која приказује куб унесеног броја.

````python
def main():
    vrednost = int(input('Za koji broj želiš ispisati kub: '))
    prikazi_kub(vrednost)

def prikazi_kub(broj):
    rezultat = broj ** 3
    print(f'Kub broja {broj} iznosi {rezultat}')

main()
````

## Прослеђивање више аргумената

Функцији се наравно може по потреби проследити више аргумената. У следећем програму приказана је функција **`prikazi_zbir`** која прихвата 2 аргумента. У функцији се сабирају та два аргумента и приказује њихов збир.

````python
# program demonstrira prosleđivanje više argumenata funkciji

def main():
    broj1 = int(input("Unesi 1. broj "))
    broj2 = int(input("Unesi 2. broj "))
    prikazi_zbir(broj1, broj2)

# funkcija prihvata dva argumenta i prikazuje njihov zbir
def prikazi_zbir(br1, br2):
    zbir = br1 + br2
    print(f'Zbir brojeva iznosi {zbir} ')

# pozivanje main funkcije
main()
````

У горњем програму видимо да су две параметарске променљиве **`br1`** и **`br2`** наведене унутар заграда у заглављу функције **`prikazi_zbir`**. То значи да ће функција прихватити вредност два аргумента.

**Ако у заглављу функције наводимо више параметара тада се они одвајају зарезима.**

Наредба у **`линији 6`** позива функцију **`prikazi_zbir`** и прослеђује јој вредност два аргумента: **`brој1`**  и **`brој2`** . Први аргумент прослеђује се првом параметру у заглављу функције, а други аргумент прослеђује са другом параметру у заглављу функције. Тако ће параметар **`br1`**  преузети вредност аргумента **`brој1`** , а параметар **`br2`**  ће преузети вредност аргумента **`brој2`**  као што је приказано на следећој слици

### Условна (while) петља

До сада смо се упознали са коришћењем *`for`* петље. Видели смо да **`for`** петљу користимо када тачно знамо колико ћемо пута одређене наредбе понављати. Због тога се for петља назива и **бројачка петља**.

Када смо се упознавали са петљама видели смо да у пајтону постоји и друга врста петље, а то је такозвана **условна петља** или **`while`** петља. **`While`** петљу користимо када не знамо колико ће бити потребно понављати наредбе у телу петље. Да ли ће се петља извршавати зависи од услова у заглављу петље. While петља је добила назив по начину како она функционише ***while a condition is true do some task*** (Док је услов тачан извршавај одређене задатке).

While петља има два дела:
* Први део је услов који се тестира или контролише. Одговор на тај услов може бити тачно или нетачно
* Други део представља низ наредби или израза који се понављају све док је услов тачан

![](slike/whilePetljaDijagram.png)

Ако је одговор на услов тачан (**`True`**), наредбе у телу петље извршавају и поново долазимо на део када проверавамо услов. Услов се поново проверава и ако је тачан понављамо читав процес. Ако је одговор на услов нетачно (**`False`**) завршава се петља и наставља се програм са првом следећом наредбом после петље.

Општи облик while петље може се приказати на следећи начин

![](slike/opstiOblikWhilePetlje.png)

У првом реду наведено је **заглавље петље**. Оно почиње речју **`while`** након којег следи **логички израз** тј. услов који ће се тестирати. Након услова наводе се две тачке. <br>
Након тога у новом реду наводе се наредбе које чине **тело петље** и оне ће се понављати у сваком кораку петље све док је услов задовољен. Наредбе које су наведене у телу петље морају бити увучене у односу на заглавље петље. Увлачењем наглашавамо које су то наредбе које чине тело петље.

Када у програму дођемо до **`while`** петље, тестира се логички израз тј. услов. Ако је услов тачан наредбе које су наведене у телу петље се извршавају. Након што се заврши последња наредба поново се тестира услов и ако је тачан наредбе у телу петље се извршавају. Након сваког проласка кроз тело петља проверава се услов. Оног тренутка када је услов нетачан петља се завршава.

Приликом упознавања **`for`** петље користили смо пример у коме израчунавамо колико ћемо имати на нашем рачуну након 5 година штедње ако су познати почетно стање и проценат годишње камате. Пошто знамо да ћемо штедети пет година за то смо користили **`for`** петљу.
Ако бисмо модификовали пример и креирали програм помоћу кога ћемо израчунати колико ће нам требати година да се стање на нашем рачуну удвостручи, морали би користити **`while`** петљу. While петљу морамо користити због тога што не знамо колико ћемо пута морати поновити наредбе у телу петље.

Програм помоћу кога бисмо исписали колико би нам година требало да удвостручимо стање на рачуну могли би анализирати овако:

![](slike/whilestednja1.png)

Почињемо тако што за годину поставимо вредност 0, а за почетно стање 10.000 динара. Понављаћемо следеће наредбе све док је стање мање од 20000 (двоструке почетне вредности)
* додавање 1 на број година
* Рачунати камату (стање * проценат камате / 100) пет процената тренутног стања
* Додавати камату на постојеће стање

На крају програма испишемо број година и стање на рачуну
Ако би овај псеудокод превели у програм он би могао да изгледа овако:

```python
# početne vrednosti
pocetnoStanje = 10000
procenatKamate = 5

stanje = pocetnoStanje
godina = 0
while stanje < pocetnoStanje * 2:
    godina = godina + 1
    kamata = stanje * procenatKamate / 100
    stanje = stanje + kamata
    print(f'Stanje nakon {godina} godina:\t{stanje:.2f}')
print(f"\nDa bismo udvostručili štednju potrebno je {godina} godina.")
print(f"Stanje na računu će biti {stanje:.2f} dinara.")
```

На почетку смо одредили променљиву **`pocetnoStanje`** и доделили јој вредност 10000. Исто тако дефинисали смо и променљиву **`procenatKamate`** и доделили јој вредност **`5`** (лакше анализирати програм када видимо назив промењиве него само број 5).

Затим смо у линији **`5`** навели да променљива **`stanje`** добије вредност почетног стања (10000). Променљивој **`stanje`** ћемо додељивари вредност тренутног стања на рачуну. <br>
Почињемио са годином која има вредност 0 (почетак штедње). 
У **`линији 7`** почиње **`while`** петља која ће се извршавати све док је стање мање од двоструког почетног стања.

У петљи понављамо следеће:
* додавање 1 на број година
* рачунамо камату
* израчунава стање тако што она тренутно стање додајемо камату

Када први пут прођемо кроз петљу број година ће бити **`1`**. Након једне године додаћемо камату и стање ће се увећати за 5%. Када смо прошли кроз све наредбе у телу петље, поново проверавамо да ли је тренутно стање које је увећано за камату и даље мање од 20.000. Ако јесте поново улазимо у петљу, повећавамо годину за један (то бити друга година), рачунамо камату (5% од тренутног стања) додајемо на стање и тако га увећавамо.

Након тога поново проверавамо да ли тако увећано стање мање од 20.000. Ако јесте улазимо поново у петљу. Све то понављамо док је услов тачан тј. док је стање мање од 20.000.

Како се стање буде увећавало сваким пролазом кроз петљу, у једном тренутку ће постати веће од 20.000 и тада услов у заглављу **`while`** петље више неће бити тачан и завршићемо петљу. Када петља заврши исписујемо колико је потребно (**`линије 12`**) и колико нам је стање на рачуну након свих тих година (**`линија 13`**)

Напишимо програм за обрачун колико купац треба да плати за производе купљене у продавници. Пошто не знамо колико производа купац има у корпи, не можемо унапред знати ни број понављања наредби у телу петље. Да бисмо израчунали колико купац треба да плати морали би понављати:

* унос цене одређеног производа
* унос количине одређеног производа
* израчунавање износа за тај производ множењем цене и количине
* додавање тог износа на укупан рачун

Када то израчунамо само за један производ то би понављали и за сваки следећи производ који је купац купио.
Ако би користили **`for`** петљу морали бисмо пребројати производе и количине и толико пута извршавати петљу. Пошто то наравно није практично, ми ћемо користити **`while`** петљу.

```python
# promenljiva kojom ćemo određivati da li
# ima još proizvoda za kupovinu
nastavak = "da"

vrednost = 0  # vrednost pojedinačnog proizvoda
ukupno = 0    # ukupan iznos za plaćanje

# petlja traje sve dok je vrednost promenljive nastavak 
# jednaka "da" (ima još proizvoda)

while nastavak == "da":
    cena = float(input("Unesite cenu: "))
    kolicina = float(input("Unesite kolicinu: "))
    vrednost = cena * kolicina
    print(f"Vrednost proizvoda: {vrednost:.2f} din.")
    ukupno = ukupno + vrednost

    nastavak = input("\nDa li ima još proizvoda? ")

print('\n' + '-' * 35)
print(f"Ukupan račun iznosi: {ukupno:.2f} dinara.")
```

Из горњег примера можемо приметити да је услов који се тестира **`dalje == 'da'`**. Петља тестира тај услов и ако је тачан наредбе од **`линије 12`** до **`линије 18`** се извршавају.

Након тога петља се наставља од **`линије 11`**. Поново се тестира да ли је вредност променљиве **`dalje == 'da'`** што значи да има још производа. 

Циклус се понавља све док одговарамо са **`da`**. Оног тренутка када унесемо другачију вредност за променљиву **`dalje`**, петља се завршава и наставља се са првом следећом наредном после петље.

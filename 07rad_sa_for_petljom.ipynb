### Рад са **`for`** петљом

Као што смо видели често се у програмима јавља потреба да се одређене наредбе извршавају неколико пута. Писање таквих наредби у пајтону је омогућено коришћењем такозваних цикличних структура или петљи.

У пајтону постоје две врсте петљи. Једна од њих је бројачка петља или **`for петља`**. For петља се назива **бројачка петља** због тога што њу користимо када знамо тачан број понављања одређених наредби тј. када тачно знамо колико пута желимо да се одређена наредба или група наредби понавља.

Претпоставимо да желимо да напишемо програм који ће рачунати укупну зараду у току једне седмице за одређену продавницу. Потребно је да у програму уносимо зараду за шест дана и на крају да испишемо укупну зараду. Уместо да шест пута пишемо наредбу за уношење одређене вредности, можемо ту наредбу написати само једном и извршити је 6 пута користећи **`for`** петљу.

У пајтону **`for`** петља функционише тако да одређена променљива, такозвана ***циклусна промењива***, добија вредности из секвенце података. Овде је приказан општи облик for петље.


```Python
for promenljiva in sekvenca:
    izraz(naredba)
    izraz(naredba)
    izraz(naredba)
    izraz(naredba)
    itd...
```

У првом реду наведено је **заглавље петље** у коме је декларисана циклусна променљива **`promenljiva.`** То је променљива коју именујемо ми у складу са правилима за именовање променљивих. Затим је наведена **`sekvenca.`** Секвенца представља низ вреднсти које ће циклусна променљива преузимати током извршавања **`for`** петље. На крају заглавља наводе се две тачке и у следећем реду или неколико редова наводе се наредбе које ће се извршавати у телу петље.

Наредба у телу петље **морају бити увучене у односу на заглавље for петље**.

**`For петља`** функционише по следећем принципу:
- циклусној променљивој је додељена прва вредност из листе
- након тога се извршавају све наредбе у телу петље.
- када се изврше све наредбе у телу петље проверава се да ли постоји још вредности у секвенци и, уколико постоји, циклусној променљивој се додељује следећа вредност и извршава тело петље.
- када више нема вредности у секвенци завршава се петља и прелази на прву следећу нарердбу после петље


Процес функционисања **`for петље`** можемо приказати следећим дијаграмом тока:
![](slike/04dijagramforpetlja.png)

Следи један једноставан програм који илуструје рад **`for`** петље:

````python
print('Ako želimo naučiti programiranje treba:')
for brojac in [1, 2, 3, 4, 5]:
    print('vežbati')
print('i što više vežbati!')
````

У <span style="color:red; font-weight:700">линији 1</span> је наваедена **`print`** функција која се извршава пре **`for`** петље.<br>
Након ње, у <span style="color:red; font-weight:700">линији 2</span> је наведено **заглавље петље** у којој се налази:
* променљива **`brojac`** - циклусна променљива која ће добијати вредности из секвенце
* **`секвенца`** $[1, 2, 3, 4, 5]$ - листа са вредностима које ће циклусна променљива добијати<br>

у <span style="color:red; font-weight:700">линији 3</span> (**тело петље**) налази се наредба која ће се понављати одређени број пута<br>
у <span style="color:red; font-weight:700">линији 4</span> наведена је **`print`** функција која се извршава када се петља заврши

Када се покрене **`for петља`** циклусној променљивој се дедељује прва вредност из листе података или секвенце **`(1)`**. У телу петље налази се само једна наредба - исписивање поруке. Када се наредба у телу петље изврши, проверава се да ли постоји још вредности у секвенци. Пошто постоји - **`вредност 2,`** циклусној променљивој се додељује вредност **`2`** и поново се извршавају наредбе у телу петље (исписује се поруке) и поново се проверава да ли постоји још вредности у секвенци. Пошто постоји, циклусна промењива добија следећу вредност - **`3`** итд.

Када циклусна променљива **`brojac`** добије и последњу вредност из секвенце - **`број 5,`** пошто не постоји више вредности у секвенци, завршава се **`for петља`** и прелази се на следећу наредбу после петље или се завршава програм.

**Вредности које се налазе у секвенци не морају бити серије бројева. Петља ће се извршавати онолико пута колико има одређених вредности у секвенци. То могу бити било који бројеви или вредности неког другог типа.**


У следећем примеру се секвенца састоји од текстуалних података (стрингова). Петља функционише потпуно исто као и када су у секвенци биле бројчане вредности. <br> У првом кораку циклусној променљивој је додељена вредност првог стринга и он се користи у телу петље. Након завршетка наредбе у телу петље, циклусној променљивој се додељује вредност другог стринга итд.<br> Пошто секвенца има 3 имена, петља ће се извршити 3 пута.

````python
for ime in ["Marko", "Jovana", "Milica"]:
    print(ime,'uči programiranje.')
````

### Коришћење **`range`** функције у **`for`** петљи

У неколико претходних примера видели смо како у заглављу петље наводимо **секвенцу** од које зависи колико ће се пута **`for`** петља извршавати. Такав начин у коме наводимо могуће вредности које ће имати циклусна променљива је погодан за мали број корака. Видели смо пример у коме пет пута исписујемо одређену поруку. Међутим, ако је потребно да се петља извршава више пута, тада је тешко наводити све могуће вредности које може имати циклусна променљива.<br> Да бисмо поједноставили процес писања бројачких петљи или **`for`** петљи, у Пајтону је омогућена коришћење **`range`** функције.<br> **`Range`** функција може да се користи на неколико начина. Следи пример који користи **`range`** функције у **`for`** петљи

````python
print('Ako želimo naučiti programiranje treba:')
for brojac in range(5):
    print('vežbati')
print('i što više vežbati!')
````

Можемо приметити да уместо листе вредности као у претходним примерима са **`for`** петљом, **`range`** функција користили аргумент **`5`**. У овом случају **`range`** функција ће генерисати секвенцу целих бројева у интервалу од **`0`** до броја **`5`** али **не укључујући број 5**.

Ако **`range`** функција има један аргумент **он одређује колико ће се пута петља извршавати** тако што циклусна променљиве креће од нуле. Када се изврше наредбе у телу петље, вредност циклусне променљиве се увећава за 1, поново се извршавају наредбе у телу петље и тако редом до вредности аргумента **`range`** функције.

Range функција може имати и више аргумената.<br>
Општи облик **`range`** функције може се приказати на следећи начин:
![Општи облик range функције](slike/06rangeFunkcija.png)

Видимо да **`range`** функција мора имати један аргумент - <span style="color:#0297d1; font-weight:700"> kraj </span> што представља крајњу вредност до које циклусна променљива добија вредности. Аргументи **`početak`** и **`korak`** нису обавезни.
Њихова примена је објашњена у наставку.

<span style="color:red"><strong>Подсећање</strong></span><br>
Ако **`range`** функција има један аргумент (нпр: **`range(10)`**) циклусна променљива добија почетну вредност **`0`** и након извршења наредби у телу петље вредност циклусне променљиве се увећава до вредности аргумента (**али не и вредности аргумента**)

</div>   
<div>

Следећи облик range функције користи два аргумента. Први аргумент означава почетну вредност, а други крајњу вредност које ће бити додељиване циклусној променљивој.
Погледајмо пример:

````python
print('Naredba pre petlje.')
for korak in range(1,6):
    print(f'{korak}. korak petlje.')
print('Naredba posle petlje.')
````

Ако покренете горњи програм видећете да ће се на екрану 5 пута исписати наредна која се налази у <span style="color:red; font-weight:700">линији 3</span>. По својим подразумеваним подешавањима range функција генерише секвенцу бројева који се аутоматски увећавају за по један од почетне до крајње вредности. И у овом случају **крајња вредност се не додељује циклусној променљивој**.

**`Range`** функција може имати и трећи облик, а то је функција са три аргумента. Прва два аргумента имају исту намену као у претходном примеру док је трећи аргумент **корак за који ће се увећавати вредност циклусне променљиве** од почетне до крајње вредности.

````python
print('Ispisivanje neparnih brojeva.')
for broj in range(1,10,2):
    print(broj)
````

У овом случају range функцији смо проследили три аргумента:
* први аргумент број **`1`** је почетна вредност у секвенци
* други аргумент - број **`10`** је крајња вредност у секвенци што значи да ће последњи број у секвенци бити број **`9`**.
* трећи аргумент - број **`2`** је корак увећања. То значи да ће се почети са вредношћу првог аргумента и даље циклусној променљивој додељивати сваки други број из секвенце бројева од 1 до 10

И још један пример:

````python
for i in range(5,101,5):
    print(i)
print('Ko se nije skrio magarac ga bio :)')
````

Ако сте се икада играли жмурке онда вам је смисао овог програма јасан :)

У свим овим наведеним примерима користили смо сукцесивно увећавање бројева у секвенци крећући се од мањег ка већим бројевима. Како можемо смањивати вредност бројева у секвенци?

Да бисмо исписали бројеве од 10 до 0 користимо следећи облик **`range`** функције:

````python
for broj in range(10,0,-1):
    print(broj)
````

Видимо да је у овом случају **први аргумент** број од којег крећемо, **други аргумент** је број до којег ћемо додељивати вредности циклусној променљивој, а **трећи аргумент** је корак и представља вредност за коју ћемо умањивати вредност циклусне променљиве док не дођемо до крајње вредности.

### Примери решавања задатака са **`for`** петљом

Из досадашњег рада са **`for`** петљом можемо закључити да се приликом осмишљавања начина на који решавамо задатак издвајају две важне ствари:

* мора нам бити јасно које наредбе ћемо понављати у петљи тј. шта је то што се понавља приликом решавања одређеног задатка
* морамо одлучити колико пута ће се наредбе у телу петље понављати или, у зависности од задатка, одакле крећемо и када завршавамо петљу (како да осмислимо заглавље петље и секвенцу у којој наводимо број понављања).

Да бисмо то правилно разумели пожељно је, пре самог рада на задатку, на папиру скицирати начин на који бисмо га решили. Треба једноставно узети папир и оловку, писати кораке које би подузимали и на основу тога закључити који се кораци понављају и колико пута.<br>

**Претпоставимо следећи задатак:**<br>
Одлучили смо да штедимо у банци и уложили смо 10.000 динара. Банка нам даје годишњу камату од 5%. Треба да испишемо колико ћемо имати новца на рачуну након пет година.

Пробаћемо решење скицирати на папиру у облику једноставне табеле:

![](slike/09tabelaStanja.png)

Овај проблем можемо решити тако што ћемо рачунати колико износи камата након сваке године и ту камату ћемо додавати на постојеће стање. То значи да би, када прође прва година штедње, камата износила 5% од 10.000 што износи 500 динара. Тај износ додајемо на стање од 10.000 и након прве године ново стање је 10.500 динара.
<br>
Када прође друга година на стање од 10.500 додајемо 5% камате што износи 525 динара и ново стање након две године постаје 11025 динара. Након треће године поново рачунамо износ камате на стање од 11.025 динара и додајемо на постојеће стање. И тако исто рачунамо камату и ново стање за четврту и пету годину.<br>

Ако посматрамо слику видимо да приликом решавања овог задатка понављамо:

* израчунавање камате (рачунамо тако што тренутно стање помножимо са процентом камате и поделимо са 100)
* тако израчунати камату додајемо на тренутно стање и то нам постаје ново стање

Те две наредбе ћемо понављати у петљи, а пошто желимо да проверимо колико ћемо имати за 5 година понављамо их пет пута.

Када смо ручно, у облику табеле, исписали на који начин ћемо добити коначно стање након пет година, сада то можемо исписати у програму:

```python
# program ispisuje stanje na računu nakon 5 godina štednje
# početno stanje je 10000 dinara, a godišnja kamata iznosi 5%

stanje = 10000
PROCENAT_KAMATE = 5  # konstante pišemo velikim slovima

for i in range(5):   # petlja će se izvršavati 5 puta
    kamata = stanje * PROCENAT_KAMATE / 100
    stanje = stanje + kamata

print(f"Nakon 5 godina na računu ćemo imati {stanje:.2f} dinara.")
```

Након извршавања овог програма видимо само коначан резултат тј. колико ћемо имати на рачуну након пет година. Да бисмо били сигурни да заиста добијамо исправне резултате најбоље је програм осмислити на папиру, а онда га превести у програмски код. Ако након извршавања програма добијамо исте резултате значи да је програм тачан.

Исто тако потребно је да програм тестирамо на познатим подацима тако да можемо проверити да ли су излазни резултати тачни.

Некада је корисно исписивати међурезултате тј. помоћу **`print`** функције исписивати тренутне вредности да можемо пратити да ли правилно рачунамо. Тако би било корисно у овом програму у сваком кораку петље исписати тренутно стање и на тај начин проверити да ли су након сваке године исти резултати и у програму и у нашој табели.

Да бисмо то урадили можемо модификовати горњи програм и исписати га на следећи начин:

```python
# Program ispisuje stanje na računu nakon 5 godina štednje
# Ispisivanje stanja se vrši nakon svake godine

stanje = 10000
PROCENAT_KAMATE = 5
print('\t\t Stanje\t   Kamata   Saldo')
for i in range(1, 6):
      kamata = stanje * PROCENAT_KAMATE / 100
      stanje = stanje + kamata
      print(f'Nakon {i}. godine: {stanje:<9.2f} {kamata:<8.2f} {stanje + kamata:.2f}')
```

У горњем примеру видимо да смо изменили и заглавље **`for`** петље тј. **секвенцу**. Сада смо користили **`range`** функцију са два аргумента, **почетак** и **крај**, да бисмо користећи циклусну променљиву могли приказивати годину и тренутно стање на рачуну.

Исто тако можемо и променити проценат камате да бисмо били сигурни да ли је то тачно.

**Корак 3: подешавање промењивих за извршавање у првом кораку петље**

Овде је корисно да испишемо све променљиве које користимо и мењамо у петљи и одредимо како да их иницијализујемо тј. како да им доделимо почетне вредности. Уобичајене вредности за бројаче и почетне вредности су најчешће 0 или 1, за збирове 0, производе 1 итд.

У нашем примеру користићемо следеће промењиве:
* тренутни дан
* тренутна температура
* највећа температура
* највећи дан

Морамо размишљати како да одредимо највећу температуру. Не можемо јој доделити вредност **`0`** (нула) пошто је могуће да уносимо све зимске температуре тако да највећа зимска температура може да буде испод нуле. <br>
Добар начин би био да прву температуру унесемо пре петље и њену вредност означимо као највећу, а највећи дан означимо као **`1`**.

Ако тако урадимо онда морамо водити рачуна о **`range`** функцији у заглавље петље. Нећемо кретати од један него од два тј. од другог дана пошто смо за први дан већ унели температуру пре петље.

Из горње табеле можемо уочити да смо првог дана унели температуру од 25 степени. Према томе највећи дан је **`1`**, а највећа температура је **`25`** степени. Затим, другог дана смо унели температуру од 24 степена и пошто она није највећа нисмо мењали највећу температуру.

Трећег дана смо унели 28 степени и пошто је она већа од највеће досадашње, највећој смо доделили ту вредност **`28`**, а највећем дану редни број дана то је **`3`**.

У четвртом дана смо унели 30 степени. Највећи дан је сада четврти, а највећа температура **`30`** пошто је она већа од највеће.

**Корак 6: писање кода у програмском језику**

Када смо написали псеудокод и када смо ручно на папиру истестирали могуће вредности, сада можемо псеудокод преписати у наредбе програмског језика. 

(12.11.2019.) П.С. Ко ово прочита и први ми се јави на мејл добаја петицу.

Програм би могао изгледати овако:

```python
# Unose se dnevne temperature za jedan mesec.
# Ispisuje se najveća temperatura i redni broj dana u kome je zabeležena

broj_dana = int(input('Koliko dana ima u mesecu: '))
temperatura = int(input('Unesi temperaturu za 1. dan: '))

# prvu unesenu temperaturu dodeljujemo promenljivoj najveca
# a promenljivoj dan (dan sa najvećom temperaturom) dodeljujemo 1
najveca = temperatura
dan = 1

# pošto smo uneli temperaturu za prvi dan
# u for petlji ćemo krenuti od 2. dana pa do poslednjeg

for i in range(2, broj_dana + 1):
    temperatura = int(input(f'Unesi temperaturu za {i}. dan: '))
    if temperatura > najveca:
        najveca = temperatura
        dan = i
print(f'Najveća temperatura je {najveca} stepeni i zabeležena je {dan}. dana.')
```
